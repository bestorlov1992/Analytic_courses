# Команда

Архитектора базы данных привлечем временно, чтобы создать хранилище.
- ETL специалист и специалист по базам данных (совмещенная должность)
    - будет следить за функционированием хд
    - поддреживать стабильную работу
    - проводить заливку в хд из разных источников
- BI аналитик
    - задача соединять работу аналитиков, датасайнтистов и бизнес
    - четкое понимание целей бизнеса
    - подготовка отчетов и дашбордов для бизнеса
    - пониание какие метрики нужны для принятия решения руководством
    - умение выбрать из множества метрик и графиков главное, нужное для принятия решения
    - отслеживание результатов принятия решений руководством и корректировка своей работы
- Аналитик
    - Глубокий анализ данных
    - Извлечение данных из хд, очистка, обработка, подготовка
    - Понимание бизнес процессов на хорошем уровне
    - проведение экспериментов, АБ тестов, проверка разнообразных гипотез
    - Простое прогнозирование
    - Рекомендации для бизнес аналитика
    - Совместное обсуждение определенных задач с дата сайнтистом
- Data Scientist
    - Максимально глубокий анализ данных, с использование машинного обучения
    - Прогнозирование разных показателей и понимание физики моделей машинного обучения
    - Подготовка заданий для аналитика, совместное обсуждение задача с аналитиком
    - Подготовка и валидация моделей  
    
Скорее всего аналитик и data scientist будет одно лицо, но бизнес аналитика лучше иметь отдельного человека.


# CRISP-DM

Задача  
В прошлом месяце значительно упали покупки билетов. Нужно выяснить в чем причина.

- Бизнес анализ
    - Проанализировать текущую ситуацию с точки зрения бизнеса
    - Определить ключевые метрики, которые влияют на покупки билетов
    - Подготовить план проекта
    - Сформулировать четкую задачу и план
- Анализ данных
    - Собрать данные по ключевым метрика, которые были выбраны
    - Понять физический смысл каждой метрики
    - Проверить качество данных, пропуски
    - Изучение данных
- Подготовка данных
    - Выбрать нужные данные
    - Почистить данные
    - Заполнить пропуски
    - Сгенеривать данные, если в этом есть необходимость
    - Соеденить данные из разных таблиц
    - Форматировать данные
- Оценка решения
    - Проанализировать полученные результаты
    - Определить в чем причина падения покупок
    - Сформировать отчет и дашборды для руководства
    - Предложить как можно восстановить продажи, основываясь на полученных данных анализа
