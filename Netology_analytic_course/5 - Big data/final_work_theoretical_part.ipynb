Вы - главный по данным в среднем по объему просмотров интернет-кинотеатре.  
Ваша задача разработать стратегию внедрения хранилища данных и работы с большими данными в этой компании.  
Задания:  

1. Описать основные бизнес-отчеты (2-3 штуки), которые мы хотим видеть по нашему бизнесу  
    - Финансовый отчет. Хотим видеть динамику прибыли в общем. Видеть расходы. Все это видеть в разрезе годов, месяцев и прочее.  
    Задача контролировать финансовую отчетность в динамике.  
    - Маркетинговый отчет. Хотим видеть динамику расходов на маркетинг и прибыль от его работы.
    - Отчет по пользователям. Задача следить в динамике за поведением пользователей. 
    Цель определить. что нужно для увеличения количества подписок. 
    - Отчет о просмотрах. Задача понимать какие фильмы лучше смотрят, на каких больше прибыли.  
    Цель определить какие фильмы лучше рекламировать и для каких групп.
2. Описать основные имеющиеся данные и источники их поступления
    Имеющиеся данные:
        - данные поведения пользователей на сайте (разные конверсии, тепловые карты поведения, и прочее)
        Источники их поступления: статистика веб сайта, разные сервисы, позволяющие получать поведение пользователей на сайте.
        - данные по просмотрам фильмов (оценки пользователей, отзывы, количество просмотров, длительность просмотров)
        Источники их поступления: страницы фильмов, где пользователи ставят оценки и отзывы, система сбора данных о просмотрах, которая записывает сколько фильм смотрели по времени, количеству раз и прочее.
        - данные данных самих пользователей (фио, почта и прочее)
        Источники их поступления: регистрационная форма
        - данные по подпискам (цены на подписки, длительность подписок)
        Источники их поступления: система по подпискам, в которой сохранятся какую подписку выбрал пользователь, ее цена, категория и прочее.
        - данные устройств (тип устройства, марка и прочее)
        Источники их поступления: системы, которые могут сохранять информацию о устройстве, с которого пользователь совершает активность на сайте
3. Описать основные сущности в хранилище данных (схема звезда) и процесс заливки данных
    Основной сущностью будет таблица подписок, так как именно она будет меняться, а на остальные таблицы она будет ссылкаться
    Таблица фактов будет содержать
    - ай ди подписки
    - ай ди клиента
    - ай ди статистики по просмотрам конкретного пользователя
    - ай ди устройства (из таблицы устройств)
    - ай ди оплаты (информация об оплате)
    - ай ди таблицы времени (информация о времени начала, окончания, продления)
    - ай ди категории (подписка возможно для разных категорий)
    - ай ди доп информации о подписке (категория и прочее)
    - сумма подписки
    Сумма подписки будет меняться, а остальные данные будут браться из соответствующих таблицы измерений
4. Описать основные проверки на качество данных (10 штук), которыми будем пользоваться при заливке
    - проверка регистрационных данные (возраст больше 0, дата рождения больше 1900 и прочие)
    - сумма подписки больше нуля
    - Конверсии на сайте больше нуля. Конверсия меньше 1. 
    - Разные даты активности на сайте больше даты регистрации на сайте
    - Длительность подписки больше 0
    - Длительность просмотра фильма меньше длинны самого фильма
    - Лубая подписка должна быть меньше или равна самой дорогой подписке на сайте
    - Поля, в которых должны быть текстовые данные должны содержать хотя бы одну букву
    - Оценки больше 0 и меньше верхней границы
    - Числовые данные, которые физически не могут быть ниже нуля, не должны быть ниже нуля
5. Придумать Data-проект, который должен улучшить показатели Вашего бизнеса и расписать его по Crisp-DM
    Data-проект - хотим предсказывать загрузку складов в разное время года
    Распишем по crisp-dm
    - Business Understanding Phase
        - Бизнес цель  
        Цель - сократить расходы на доставку и хранение товаров, которые пользуются малым спросом в разное время года  
        Болью бизнеса в данной ситуации являются расходы на хранение, доставку товаров. Мы замораживаем деньги в товары, которые не покупают. Хотим уменьшить это.  
        Критерием успеха будет сокращение денег, направленных на разные товары, которые не пользуются спросом в этот период времени.  
        - Текущая оценка систем  
        Ключевые пользователи - люди, которые отвечют за респределение денег (фин отдел, менеджеры по закупкам, директор по развитию)
        Результат должен быть представлен в виде модели, которая будет определять количество товаров, которые будут востребованы в следующем периоде.
        - Декомпозиция задач  
        Задача разбивается на следующие подзадачи
            - Собрать данные
            - Почистить данные
            - Определить ключевые показатели, которые будут использоваться для модели
            - Выбрать модель для задачи
            - Обучить моедль и подобрать макропараметры
            - Оценить результаты
            - Реализовать возможность получать прогноз на следующий месяц в удобной форме
    - Data Understanding Phase
        - Сбор данных
        Источником данных будет хранилище данных, в которое сохраняются данные о покупках
        - Описание данных
        Определяем качество данных. Сколько пропусков, насколько каждый параметр не выходит за допустимые физические значения. 
        Пнять причину пропусков или некорректных данных. 
        Заполняем пропуски или удаляем данные, если заполнение приведет к некорректным результатам.  
        Вычисляем базовые статистики для необходимых параметров, чтобы лучше их изучить
        - Исследование данных
        Строим одномерные графики, чтобы понять распределение параметров.  
        Изучаем взаимодействие атрибутов. Для этого строим матрицу корреляции и скэтэрплоты для пар параметров.
        Определить какие параметры коррелируют между собой и какие коррелируют с целевой переменной.  
        Определить виды распределений разных параметров.
    - Data Preparation Phase
        - Интеграция данных  
        Объединаяем данные из разных таблиц в одну.  
        Если нужно, то аггрегируем данные по каким то группам.
        - Выбор данных
        Выбираем атрибуты для модели.  Описываем почему мы выбрали именно эти атрибуты. И почему не взяли другие.  
        Разбиваем данные на подвыборки - тестовую, проверочную и обучающую.
        - Преобразование данных  
        Снижение размерности  
        Нормализация, дискретизация, бинаризация  
        Построение данных (Feature Engeniring). Важный пункт, так как часто из существующих можно вычленить дополнительные атрибуты или создать новые, например, с помощью разбиения на категории (дискретизация)
    - The Modelling Phase  
        - Выбор модели
        Выбираем модель в зависимости от имеющихся атрибутов, их распределений, особеностей параметров. В зависимости от необходимых результатов.  
        Перебираем подходящие модели.  
        - Разработать режим тестирования
        Определиться как будем тестировать.  
        Убедиться, что обучающая, проверочная и тестовая выборки имеют сходные характеристики.  
        Убедиться. что выборки репрезентативны.  
        - Построение модели
        Выбираем начальные макропараметры.  
        Исследуем поведение модели, анализируем как она реагирует на изменения в данных, устрйчивость.  
        - Оценка модели
        Избегать переобучения.  
        Определить сегменты данных, где модель менее эффективна. Определить причины этого.  
        Важный пункт, так как он покажет нам, при каких данных модель может ошибаться больше обычного.  
    - The Evaluation Phase
        - Валидируем модель
        Оцениваем результаты с точки зрения бизнеса. Определяем устраивает ли качество бизнес.  
        Достигается ли цель моделью.  
        Определяем ожидаемые результаы (ROI и другие)
    - The Evaluation Phase
        Оценить потенциал от внедрения.  
        Описать возможные проблемы.  
        Описать ситуации в которых модель лучше хуже предсказывает, в зависимости от данных и от внешних обстоятельств.
    - The Deployment Phase
        Схема внедрения  
        Презентация модели  
        Интеграция с существующей инфрастуктурой  
        Генерация отчетов
        Описать подробный отчет по внедрению с максимальной детализацией.  
6. Описать требуемые роли в команде по работе с данными на этапах 4 и 5
    Бизнес-аналитик определяет цели бизнеса для нашего проекта. Его задача связать цель с бизнесом, чтобы в итоге бизнес получил максимум пользы от модели.  
    Data Scientist извлекает данные из хранилища, исследует их, готовит для модели. Стоит модель и анализиует результаты.
    Работу Data Scientist в зависимости от объема данных и сложности можно разделить между ним и аналитиком.
    Когда модель готова, Data Scientist вместе с бизнес аналитиком оценивают насколько модель удовлетворяет поставленной бизнес цели.
    Дальше подключаются ай ти специалисты, для реализации удобного представления работы модели (софт, отчеты и прочее). Чтобы те, кто будут принмать решения на основе модели, могли это делать максимально удобно.
