Накидать 5-6 кейсов применения больших данных и анализа данных для крупного сетевого строительного магазина.

1. Оптимизировать загрузку складов. Чтобы товары, которые долго лежат не занимали место. И лучше хранить меньше, а освободившееся место отдать  
для быстро раскупающихся товаров.  
Метрики:
    - время нахождения товара на складе
    - объем места на складе (чем больше товар, тем больше убытка от его простоя)  
    - оборачиваемость товара, то есть сколько раз он поступает на склад и уходит со склада
2. Оптимизировть затраты на доставку товаров до складов и со складов. То есть нам нужно оптимально выбирать в какой из складов ближе вести новый товар  
от поставщика, если есть одинаковые скады. И что более важно, с какого склада лучше доставлять товар покупателям.  
Тут важно учесть не только физическое растояние, но и загрузку складов, количество оставшегося товара и прочее.  
Так как если у нас есть 2 склада, то может быть выгоднее доставить с более дальнего склада, так как там больше излишек и не нужно будет доставлять новый товар после  
после продажи.  
Метрики:
    - Время доставки до склада от поставщика
    - Время доставки со склада до покупателя
    - Объем доставки (возможно выгоднее отвезти с другого склада, если там большой заказ в том же направлении, пусть подальше, но поедет одна машина)
3. Оптимизация запасов в зависимости от сезонности. Задача уменьшить издержки на хранение не пользующихся спросом товаров не в сезон  
Метрики:
    - ARPU (сравниваем выручку на разные виды товаров в разные сезоны)
    - Количество покупателей для разных товаров в зависимости от времени года
4. Оптимизировать расположение товаров в магазине, чтобы определить какие товары лучше расположить ближе ко входу  
Метрики:
    - ARPU
    - количество покупок товаров в зависимости от их позиции (то есть узнать каких товаров болше покупают, если они ближе ко входу)
5. Определить в каких магазинах нужно деражть больше или меньше машин для доставки  
Метрики:
    - Количество заказов с доставкой
    - Средний размер доставки
6. Определить загружженость магазинов и оптимизировать график работы персонала (кассиры, грузчики, продавцы и другие)  
Метрики:
    - среднее количество покупателей в зале
    - ARPU в разные промежутки дня
    - среднее количество покупающих покупателей в зале
