### Сбор требований и вопросы к Заказчику <a class="anchor" id="requirements"></a>

1. Кто является заинтересованными участниками (стейкхолдерами), кто будет пользоваться результатом работы? (желательно узнать ответы всех стейкхолдеров)
2. Как вы видите решение задачи? Что мы будем решать? Какой результат ожидаете?
Обсудить что хотели бы видеть. Этот пункт очень важен. По сути это и есть задача.
Какое соотношение подозрительных и уволившихся будет приемлемо. Тут лучше обозначить рамки разных уровней. То есть идеально, нормально и не устраивает.   
Так как точных рамок естественно  не будет. Чтобы мы могли понимать, что плохо, куда стремиться.
3. Что конкретно не нравится в предыдущем решении. Пользовались ли вы им практически. Какие минусы и плюсы можете отметить.  
Тут можно понять что заказчик ждет, так как он может в предыдущем пункте не указать, но вот про предыдущий опыт сказать, что понравилось.
Больше открытых вопросов, которые могут ответить на другие вопросы. Нужно стараться избегать вопросов, на которые будет ответ да или нет.
4. Какие данные имеются? Какие метрики доступны? Можно ли получить дополнительные данные, если они будут нужны?  
Тут важно понять какими ресурсами мы распологаем. Возможно какие-то данные мы можем дополнительно получить через дополнительные каналы.
Вот тут важно понять какие есть вообще данные. И также выделить те, которые непосредственно нужны для решения задачи. Не нужно брать все даныые.
Вообще после постановки цели, понимания чего хочет заказчик, далее исходим из этого. Не тратим время на лишние данные, которые нам не понадобятся.
И наоборот стараемся получить данные, которые будут влиять на решение.
5. Определить сроки решения задачи. Лучше сразу увелить их в 1.5 - 2 раза. Так как на практике будет дольше, чем в теории. Нужно объсянить заказчику, что это увеличит эффективность результата.

 

### Ответы на вопросы Заказчика и рассказ о проекте <a class="anchor" id="answers"></a>

Этот пукнт важен тем, что тут мы можем вовлечь заказчика в процесс. Заинтересовать его. А это улучшит нашу работу.  
Поэтому важно расскзать как мы видим решенеие проблемы, что было бы желательно для решения и прочее.  
Но тут важно придерживаться цели и задачи, которую озвучил заказчик. Мы не ставим цели и не предлагаем свои цели.  
Мы на словах объясняем как мы видим решение и далее заказчик может дополнить свои задачи и лучше вовлечься в процесс.
1. Что принесет наше решение проблемы. Важно расскзать в терминах заказчика.
Благодаря нашему решению, мы сможем сузить крук потенциальных увольняющихся. 
Вы сможете в реальном времени видеть какие показтели влияют на то уволиться ли человек, и принимать привентивные меры. 
2. Теперь нам нужно обрисовать моменты, которые помогут добиться цели.
Какие дополнительные данные можно получить. Тут многое будет зависеть от имеющихся и от картины в целом.
3. Обозначим примерные сроки выполнения.
4. На основе чего мы будем пониматЬ.что результат верный или не верный. 
5. Стараемся рассказать о том, что будет после решения, какие возможны сценарии развития и что может улучшить результат.

### Обсуждение формата результат <a class="anchor" id="howitwillbe"></a>

В каком виде хотите увидеть решение? Например, просто список сотрудников с вероятностью их возможного ухода, отсортированный.  
Или дополнительно хотите увидеть факторы, которые сильнее других влияют на увольнение. Эти факторы можно вывести на основной дашборд.  
Можно дрил даун. Набросать чертеж. Какие графики больше нравятся. 

### План отчета <a class="anchor" id="reportplan"></a>

1. Введение
Во введении желательно в свободной форме описать ситуацию. Какие факторы влияют на ситуацию.  
Важно понимать, что это не постановка задачи. Это как бы описание ситуации в которой мы находимся.
И что влияет на теущее состояние проблемы. Чем больше мы сможем понять о проблеме, тем проще будет дальше.
С чем будем работать. Есть сотрудники. Кто-то может уволиться.  
На сотрудника могут влиять разные факторы. Кому-то не нравятся условия работы. Кто-то хочет больше зп.
Кому-то не нравится коллектив. Возможно нужно изучить насколько дружное общение и прочее. На решение человека могут влиять разные факторы.
Растояние до работы тоже важный фактор. Также по возможнсти было бы хорошо понимать состояние человека в семье, в коллективе и прочие факторы, которые оказывают на психологическое состояние человека. Нас интересуют характиристики, которые мы можем измерить. 
2. Постановка задачи. Ожидаемый результат.
Тут многое будет зависить от общещния с заказчиком.
Наша задача максимально точно определить кто из сотрудников уволиться. Заранее за 3 месяца.
Результатом будет моделька, которая выдает вероятности того, что человек уволиться. 
Результат представим в виде дэшборда. На верхнем уровне расположим список самых вероятных сотрудников на увольнение.
Выведем 10 самых вероятных сотрудников отсортированных по убыванию. 
Возле каждого сотрудника будет индикатор, чтобы можно было легко граффически понимать насколько он может уволиться. Не только числовое.
Возможно будет удобнее сделать круговой индикатор.
Также на верхнем уровне выведем главные факторы, которые влияют на увольнение. Именно числовые факторы, которые присутствуют в модели.
Для каждого фактора, от которых зависит увольнение сделаем drilldown. Чтобы можно было нажать на фактор и оказаться на уровне, где этот фактор стал главным и мы   
видим другие факторы, которые влияют уже на него. Резултат также оформим в виде индикаторов.
В данной задачи нам графики сильно не нужны, так как нам нужно видеть индикативно какое состояние того или иного фактора.
3. Описание данных.
Тут будет описание данных которые мы будем использовать.
Этот этап больше технический. Но важно максимально описывать словами выводы. Чтобы потом вернуться и понить что-то дополнительно.
Также каждый блок должен содержать результат его работы в словестной форме простыми словами. Чтобы заказчик мог прочитать, елси захочет углубиться.
Важно декомпозиция факторов и метрик. Нужно понять какие факторы от чего зависят, построить иерархию метрик. Это будет нужно для drill down.
Описываем источники из которых будем получать данные. Описываем почему выбрали именно эти параметры. 
    * Чистим данные. Заполняем пропуски. Готовим данные для анализа. Выбираем зависимую переменную, предикаты.
    * Изучаем каждую переменную в отдельности. Выбросы. Распределение. Что вообще физически эта переменная из себя представляет.  
    Проверяем соответствуют ли реальные данные физическим границам и прочее. Описываем каждую переменную.
    * Далее выбираем количественные переменные и смотрим как они коррелируют. Пытаемся понять какие переменные нам нужны для модели, а какие будут не нужны. 
    * Далее попарно смотрим отдельно как взаимодействуют переменные, которые мы выбрали для модели, с зависимой переменной.
    * Также проводим статистические тесты, чтобы понять как разные факторы влияют на те или иные переменные. 
В итоге будет много графиков и текста. Нужно описать что мы видим. Тут не нужно думать о целой задаче. Нам нужно сфокуссироваться на каждой переменной, на взаимодействии переменных. Задача этого этапа понять какие переменные влияют на увольнение уже в цифрах. Пощупать данные. Увидеть графически что происходит. Как распределены переменные и прочее. 
4. Выбор и подготовка переменных для моделирования.
По результатам предыдущего этапа мы должны иметь представление о переменных.
Теперь наша задача отобрать переменные для модели. Тут можено написать почему мы выбрали эти переменные. 
5. Описание моделирования и тестирования. Описание сценариев. Описание результатов.
Далее мы проводим моделирвоание. Тестируем разные параметры. Разные модели. Каждый результат протоколируем. Выводим ошибки, результаты.
Описываем результат работы разных моделей. Минусы, плюсы. Словами делаем выводы по каждой модели.
Далее мы уже выбираем лучшую модель, описываем почему выбрали эту модель. Также описываем что можнобудет ещё улучшить, если оплучиться.
В этом пункте наша задача протестировать разные модели, найти саму оптимальную. 
6. Оформление результата.
Формируем отчет. По сути это будет отчет, что мы делали выше. Все наши описания мы скрываем в ячейках. Каждый пункт будет содержрать краткое описание результата на этом этапе. В простой форме. Это нужно. если заказчик захочет деталеьнее пощупать резульататы. Ну и также будет возможность drill down. Если захочет, может раскрыть ячейку и посомтреть графики и прочее. Но детальная информация это больше для нас самих.
Формируем дэш борд. В идеале сделать обновление информации на дашборде. чтобы скприпт поттаскивал актуальную инфу и выводил на экран обновленные данные.
7. Заключение. 
Здесь описываем возможные сценарии работы с дэшбордом. И описание возможных сценариев взаимодействия с ним.
Наша задача сделать не черный ящик. Наша задача, чтобы заказчик начал доверять дэшборду. Чтобы мог принимать решения на его оснвое.
Для этого мы в отчете простыми словами описывали каждый этап. чтобы у заказчика сложилось понимание как все работает.
