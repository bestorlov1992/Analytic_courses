Выбрать любой-интернет магазин и на основе теории AARRR продумать все важные метрики бизнеса, за которыми вы как собственник стали бы следить в первую очередь.

Рассмотрим, например, интернет-магазин продажи спорт товаров Спортмастер
Рассмотрим какие метрики можно создать на основе фреймворка AARRR
* Acquisition
    * количество посещений сайта по сравнению с предыдущими периодами
* Activation
    * количество новых регистраций
    * время на сайте в первый раз
    * глубина просмотра стартовой страницы, глубина по сайту
    * положительный отклик в первое же посещение
* Retention
    * сколько повтороно приходят людей
    * сколько повторно покупают
    * любая акитвность на сайте
* Revenue
    * средний чек за месяц, день в абсалюте и относительно предыдущих периодов
    * если есть реклама на сайте, то сколько мы получили от переходов по этим ссылкам
* Refferal
    * репосты
    * ссылки в соц сетях
    * все что позволяет нам определить долю новых пользователей в 1 пункте от старых пользователей, то есть сколько людей пришло благодаря тем, кто уже был раньше на сайте
