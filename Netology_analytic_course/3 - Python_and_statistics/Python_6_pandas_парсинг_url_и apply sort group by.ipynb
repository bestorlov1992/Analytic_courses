```python
# оборачиваем все в функцию
# в качестве аргумента будет строка датафрейма

def campaign_name(row):
    """Получение названия кампании из ссылки внутри строки row"""

    parsed = parse.urlsplit(row['link'])
    params_dict = parse.parse_qs(parsed.query)

    return params_dict['utm_campaign'][0]
```

```python
stats['campaign'] = stats.apply(campaign_name, axis=1)
stats.head()
```

```python
# как передать несколько аргументов

def power_up(row, n):
    """Возводит значение столбца effect в степень n"""
    return row['effect'] ** n
```

```python
stats['power_up'] = stats.apply(power_up, n=3, axis=1)
stats.head()
stats = stats[
    ['link',
    'group',
    'phrase',
    'effect',
    'ad_id',
    'title',
    'text',
    'word_count',
    'power_up']]
stats.head()
```

Т. е. если поисковый запрос содержит название города региона, то в столбце 'region' пишется название этого региона. Если поисковый запрос не содержит названия города, то ставим 'undefined'.

Правила распределения по регионам Центр, Северо-Запад и Дальний Восток:

geo_data = {

    'Центр': ['москва', 'тула', 'ярославль'],

    'Северо-Запад': ['петербург', 'псков', 'мурманск'],

    'Дальний Восток': ['владивосток', 'сахалин', 'хабаровск']

}

Результат классификации запишите в отдельный столбец region.

```python
kw = pd.read_csv("keywords.csv")
```

```python
geo_data = {

    'Центр': ['москва', 'тула', 'ярославль'],

    'Северо-Запад': ['петербург', 'псков', 'мурманск'],

    'Дальний Восток': ['владивосток', 'сахалин', 'хабаровск']

}
def func(x):
    for key, el in geo_data.items():
        for city in el:
            if city in x.keyword:
                # print(pd.Series(key))
                return pd.Series(key)
    else:
        return pd.Series("undefined")
kw["region"] = kw.apply(func, axis=1)
kw.head()
```

Напишите функцию, которая классифицирует фильмы из материалов занятия по следующим правилам:
    - оценка 2 и меньше - низкий рейтинг
    - оценка 4 и меньше - средний рейтинг
    - оценка 4.5 и 5 - высокий рейтинг

Результат классификации запишите в столбец class

```python
import pandas as pd
ratings = pd.read_csv("ml-latest-small/ratings.csv")
movies = pd.read_csv("ml-latest-small/movies.csv")
ratings.head()
```

```python
def func(el):
    if el <= 2:
        return "low rating"
    elif el <= 4:
        return "middle rating"    
    else:
        return "high rating"
ratings["class"] = ratings.rating.apply(func)
ratings.head()
```

Посчитайте среднее значение Lifetime киноманов (пользователи, которые поставили 100 и более рейтингов). Под Lifetime понимается разница между максимальным и минимальным значением timestamp для каждого пользователя. Ответ дайте в днях.



```python
def func(row):
    days = (row["max"] - row["min"]) / 86_400
    return pd.Series(days)
res = (ratings[["userId", "timestamp"]]
    .groupby(by="userId")
    .agg({"userId" : ["count"], "timestamp" : ["min", "max"]})
    .reset_index()
)
res = res[(res["userId", "count"] >= 100)]
res["Lifetime"] = (res["timestamp"]
    .apply(func, axis=1)
    .head()
)
res.head()
```


Есть мнение, что "раньше снимали настоящее кино, не то что сейчас". Ваша задача проверить это утверждение, используя файлы с рейтингами фильмов из материалов занятия. Т. е. проверить верно ли, что с ростом года выпуска фильма его средний рейтинг становится ниже.

При этом мы не будем затрагивать субьективные факторы выставления этих рейтингов, а пройдемся по следующему алгоритму:

1. В переменную years запишите список из всех годов с 1950 по 2010.

2. Напишите функцию production_year, которая каждой строке из названия фильма выставляет год выпуска. Не все названия фильмов содержат год выпуска в одинаковом формате, поэтому используйте следующий алгоритм:
    - для каждой строки пройдите по всем годам списка years
    - если номер года присутствует в названии фильма, то функция возвращает этот год как год выпуска
    - если ни один из номеров года списка years не встретился в названии фильма, то возвращается 1900 год

3. Запишите год выпуска фильма по алгоритму пункта 2 в новый столбец 'year'

4. Посчитайте средний рейтинг всех фильмов для каждого значения столбца 'year' и отсортируйте результат по убыванию рейтингай на его выполнение)

```python
def func(el):
    years = range(1950, 2011)    
    for year in years:
        if str(year) in el:
            return year
    else:
        return 1900
movies["year"] = movies.title.apply(func)
joined = ratings.merge(movies, how="left", left_on="movieId", right_on="movieId")
joined.groupby("year").agg({"rating" : "mean"}).reset_index().sort_values("rating", ascending=False)
```
