```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('data.csv', header=0, names=['time', 'user', 'action', 'call_place', 'product_version'])
df.head()
```

1. Узнайте, какими кнопками пользуется больше всего людей. Используйте тот же набор данных.

```python
df1 = df[df.call_place == 'Панель инструментов']
```

```python
unique_users = df1.user.nunique()
unique_users
```

```python
df1_grouped = df1.groupby('action')[['user']].nunique().sort_values('user', ascending=False).apply(lambda x: round(x / unique_users * 100, 2))
df1_grouped.head(10)
```

2. Узнайте, какие пункты меню имеет смысл вынести на панель с иконками. Используйте тот же набор данных.

```python
df2 = df[df.call_place == 'Главное меню']
unique_users = df2.user.nunique()
unique_users
```

```python
df2_grouped = df2.groupby('action')[['user']].nunique().sort_values('user', ascending=False).apply(lambda x: round(x / unique_users * 100, 2))
df2_grouped.head(10)
```

```python
df2_grouped[~df2_grouped.index.isin(df1_grouped.index)].head(10)
```
