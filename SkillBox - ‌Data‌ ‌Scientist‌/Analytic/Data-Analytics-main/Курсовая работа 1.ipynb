```python
#Была первая домашка
first_hw_list = dt_main[dt_main['is_homework']].groupby(['course_title','module_number']).\
                agg({'lesson_number':'min'}).reset_index().\
                sort_values(['course_title','module_number']).\
                groupby(['course_title']).first().reset_index()
first_hw_list
```

```python
filter = dt_main.set_index(['course_title','module_number','lesson_number']).\
         index.isin(first_hw_list.set_index(['course_title','module_number','lesson_number']).index)
second_task = dt_main[filter & dt_main['is_homework'].isin([True])].groupby(['student_id','course_title'])['start_date'].aggregate(start_date='min').reset_index()
second_task['start_date'] = pd.to_datetime(second_task['start_date'])
second_task['start_month'] = pd.DatetimeIndex(second_task['start_date']).month
second_task['start_year'] = pd.DatetimeIndex(second_task['start_date']).year
second_task = second_task.groupby(['course_title','start_year','start_month']).agg(start_date=('start_date','min'),students_number=('student_id','count')).reset_index()
second_task['day'] = second_task['start_date'].dt.strftime('%Y-%m')
second_task
```

```python
student_courses = dt_main.groupby(['course_title'])['student_id'].apply(pd.unique).apply(list)
```

```python
final_table = pd.DataFrame({'Period':pd.date_range(min(second_task['start_date']), 
                                               max(second_task['start_date']),
                                               freq='MS').strftime('%Y-%m')})
for i in student_courses.index:
    final_table[i] = 0
final_table.index = final_table.Period
for i in range(second_task.shape[0]):
    final_table.loc[second_task['day'][i],second_task['course_title'][i]] = second_task['students_number'][i]
final_table = final_table.drop('Period',axis=1).sort_index().fillna(0)
final_table
```

```python
for i in student_courses.index:
    temp = final_table.loc[final_table[i].where(final_table[i]>0).first_valid_index():,i]
    temp.plot(kind='line',title=i,xlabel='Время начала',ylabel = 'Кол-во новых студентов',rot=90,grid=True)
    plt.xticks(range(len(temp.index)), temp.index)
    plt.show()
```

Для улучшения качества контента полезно выявить причину. На основании имеющихся данных придумайте метрику успеваемости студента. **Обоснуйте ее.** Выявите таких студентов на каждом курсе, чтобы предоставить продюсерам список проблемных студентов. 

```python
df_main_4 = dt_main.copy()
 
df_main_4['done'] = df_main_4['status']=='done'
df_main_4['done'] = df_main_4['done'].astype(int)

df_main_4 = df_main_4.groupby('student_id').agg({'status':'count', 'done':'sum'}).reset_index()

df_main_4['perf'] = df_main_4['done']/df_main_4['status']

df_main_4.sort_values(by=['perf','status'],  ascending=[True, False])
```
