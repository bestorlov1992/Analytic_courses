```python
df['diff'] = (df.Close - df.Open) * 100 / df.Open
color = df['diff'].apply(lambda x: 'green' if x >= 0 else 'crimson').values
fig = px.bar(df, y='diff')
fig.update_traces(marker_color=color)
fig
```
