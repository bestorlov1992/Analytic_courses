### 1.10
Для каждого классификатора выведите матрицу ошибок и самостоятельно рассчитайте метрику recall (расчеты должны присутствовать). Проверьте расчеты, воспользовавшись встроенной метрикой recall_score.

```python
from sklearn.metrics import confusion_matrix, recall_score
```

```python
### YOUR CODE HERE ###
def matrix_error(y_test,y_predicted):
    tp = ((y_test == y_predicted)&(y_predicted==1)).sum()
    tn = ((y_test == y_predicted)&(y_predicted==0)).sum()
    fp = ((y_test != y_predicted)&(y_predicted==1)).sum()
    fn = ((y_test != y_predicted)&(y_predicted==0)).sum()
    return {'tp':tp, 'tn':tn,'fp':fp,'fn':fn}

```

```python
knn_predicted = knn_model.predict(x_test)

gnb_predicted = gnb_model.predict(x_test)

dtc_predicted = dtc_model.predict(x_test)

lg_predicted = lg_model.predict(x_test)

print('knnn\b')
gnb_errors = matrix_error(y_test,gnb_predicted)
print(gnb_errors)
gnb_recall = gnb_errors['tp']/(gnb_errors['tp'] + gnb_errors['fn'])
print('recall',gnb_recall)
gnb_lib = recall_score(y_test,gnb_predicted)
print('lib',gnb_lib,'\n------')

print('Bayes n\b')
knn_errors = matrix_error(y_test,knn_predicted)
print(knn_errors)
knn_recall = knn_errors['tp']/(knn_errors['tp'] + knn_errors['fn'])
print('recall',knn_recall)
knn_lib = recall_score(y_test,knn_predicted)
print('lib',knn_lib,'\n------')

print('Tree n\b')
dtc_errors = matrix_error(y_test,dtc_predicted)
print(dtc_errors)
dtc_recall = dtc_errors['tp']/(dtc_errors['tp'] + dtc_errors['fn'])
print('recall',dtc_recall)
dtc_lib = recall_score(y_test,dtc_predicted)
print('lib',dtc_lib,'\n------')

print('Log error n\b')
lg_errors = matrix_error(y_test,lg_predicted)
print(lg_errors)
lg_recall = lg_errors['tp']/(lg_errors['tp'] + lg_errors['fn'])
print('recall',lg_recall)
lg_lib = recall_score(y_test,lg_predicted)
print('lib',lg_lib,'\n------')

```
